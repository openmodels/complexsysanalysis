## A bit about constructing a list of all the primes to 100

```r
1:100
```

```r
nn = 53
prime <- T
for (ii in 2:52) {
  if (nn %% ii == 0)
    prime <- F
}
```

```r
prime
```

```r
is.prime <- function(nn) {
  prime <- T
  for (ii in 2:(nn-1)) {
    if (nn %% ii == 0)
      prime <- F
  }
  prime
}
```

```r
is.prime(53)
```

```r
is.prime(51)
```

```r
primes <- c()
for (xx in 2:100) {
    if (is.prime(xx)) {
        primes <- c(primes, xx)
    }
}
```

```r
plot(primes)
```

# Regime shifts

1. What are regime shifts?


Let's grab a (very little) data from `Positive feedbacks promote power-law clustering of Kalahari vegetation (Scanlon et al. 2007)`
 - Kalahari Transect in Southern Africa
 - Canopy fractional tree cover ($f_t$) predicted by mean season rainfall ($\bar{r}$).
 
 

1. Look at the data.
2. Take a systems view of regime shifts.
3. Take a complex systems view of regime shifts.

## 1. Look at the data

```r
df <- data.frame(ft=c(.65, .54, .32, .19, .14, .04),
                 rbar=c(879, 811, 698, 409, 365, 216))
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + geom_smooth(method='lm') +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
ggplot(df, aes(rbar, ft, group=rbar > 600)) +
    geom_point() + geom_smooth(method='lm') +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + geom_smooth() +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

### 1a. Reproduce the basic regressions

```r
lm(ft ~ rbar, data=df)
```

```r
summary(lm(ft ~ rbar, data=df))
```

```r
mod = lm(ft ~ rbar, data=df)
```

```r
mod$coeff
```

```r
projdf = data.frame(rbar=c(0, 1000))
projdf$ft = mod$coeff[1] + mod$coeff[2] * projdf$rbar
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + geom_line(data=projdf) +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
predict(mod, projdf)
```

```r
predict(mod, projdf, interval='conf')
```

```r
projdf = data.frame(rbar=c(0, 1000))
projdf = cbind(projdf, predict(mod, projdf, interval='conf'))
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + geom_line(data=projdf, aes(y=fit)) + geom_ribbon(data=projdf, aes(y=fit, ymin=lwr, ymax=upr), alpha=.5) +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
projdf = data.frame(rbar=0:1000)
projdf = cbind(projdf, predict(mod, projdf, interval='conf'))
```

```r
ggplot(df, aes(rbar, ft)) +
    geom_point() + geom_line(data=projdf, aes(y=fit)) + geom_ribbon(data=projdf, aes(y=fit, ymin=lwr, ymax=upr), alpha=.5) +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

# 3. A complex systems view

Generate more representative data.

```r
rbars <- runif(1000, 200, 900)
fts <- sapply(rbars, function(rbar) {
    probs <- dnorm(rbar - df$rbar, 0, 200)
    main <- sample(df$ft[c(1, 6)], 1, prob=probs[c(1, 6)])
    (.6 * mean(sample(df$ft[2:5], 5, replace=T, prob=probs[2:5])) + .4 * main)
})
```

```r
df2 <- data.frame(ft=fts, rbar=rbars)
```

```r
ggplot(df2, aes(rbar, ft)) +
    geom_point() + theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
ggplot(df2, aes(rbar, ft)) +
    geom_point() + geom_smooth() +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```

```r
ggplot(df2, aes(rbar, ft, group=ft > .34)) +
    geom_point() + geom_smooth(method='lm') +
    theme_bw() + xlab("mean wet season rainfall") + ylab("fractional tree cover")
```
